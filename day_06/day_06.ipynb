```
import pandas as pd
import seaborn as sns
```

# **Tips data**

```
df=pd.read_csv('/content/tips.csv')
df
```

# **1. Scatterplot (Numrical-Numerical)**

```
sns.scatterplot(x=df['total_bill'], y=df['tip'],hue=df['sex'] ,style=df['smoker'],size=df['size'])
```

# **2. BAR Plot (Numercal-Categorical)**

```
af=pd.read_csv('/content/train.csv')
af
```

```
sns.barplot(x=af['Pclass'], y=af['Age'], hue=af['Sex'])
```

# **3. Box Plot (Numercal-Categorical)**

```
sns.boxplot(x=af['Sex'],y=af['Age'],hue=af['Survived'])
```

# **4. Distplot (Numercal-Categorical)**

```
sns.distplot(af['Age'])
```

```
sns.distplot(af[af['Survived']==0]['Age'])
sns.distplot(af[af['Survived']==1]['Age'])
```

```
sns.distplot(af[af['Survived']==0]['Age'],hist=False)
sns.distplot(af[af['Survived']==1]['Age'],hist=False)
```

# **5. HeatMap (Categorical-Categorical)**

```
pd.crosstab(af['Pclass'],af['Survived'])
```

```
sns.heatmap(pd.crosstab(af['Pclass'],af['Survived']))
```

```
(af.groupby('Embarked').mean(numeric_only=True)['Survived']*100)
```

```
(af.groupby('Embarked').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

# **6. ClusterMap(Categorical-Categorical)**

```
pd.crosstab(af['SibSp'],af['Survived'])
```

```
sns.clustermap(pd.crosstab(af['SibSp'],af['Survived']))
```
