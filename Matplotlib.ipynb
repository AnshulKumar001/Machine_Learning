```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y)  #sine wave
plt.show()
```

```python
plt.plot(x,z)  #cosine wave
plt.show()
```

# **Adding Title, x-axis and y-axis labels**

```python
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine value')
plt.title('Sine Wave')   #Title
plt.show()
```

# **parabola**

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')   #color and represent symbol
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'b--')
plt.show()
```

# **Bar Plot**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['C','C++','Java','Python','PHP']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

# **Pie chart**

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
langs=['C','C++','Java','Python','PHP']
students=[23,17,35,29,12]
ax.pie(students,labels=langs,autopct='%1.1f%%')
plt.show()
```

# **scatter plot**

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
plt.scatter(x,y,color='r')
plt.scatter(x,z,color='g')
plt.show()
```

# **3D scatter**

```python
fig2=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
