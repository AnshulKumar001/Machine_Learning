# **Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load a dataset
df = sns.load_dataset('iris')

# Calculate correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Feature Correlation Heatmap")
plt.show()
```

# **pairplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset lete hain
df = sns.load_dataset('iris')

# Pairplot bana rahe hain
sns.pairplot(df, hue='species', diag_kind='kde')

plt.suptitle("Iris Dataset Pairplot", y=1.02)  # Thoda title add kar diya
plt.show()
```

# **Boxplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset load kar rahe hain
df = sns.load_dataset('iris')

# Boxplot bana rahe hain for petal length by species
sns.boxplot(x='species', y='petal_length', data=df)
plt.title("Boxplot of Petal Length by Species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = sns.load_dataset('iris')

# Histogram plot bana rahe hain for petal length
sns.histplot(data=df, x='petal_length', kde=True, bins=20, color='purple')
plt.title("ðŸŒ¸ Petal Length Distribution")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
df = sns.load_dataset('iris')

# KDE plot bana rahe hain for petal length
sns.kdeplot(data=df, x='petal_length', fill=True, color='orange', linewidth=2)

plt.title("ðŸŒ¼ Petal Length KDE Plot")
plt.xlabel("Petal Length")
plt.ylabel("Density")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = sns.load_dataset('flights')

# Line plot for number of passengers over months
sns.lineplot(data=df, x='month', y='passengers', marker='o', color='green')

plt.title("ðŸ›« Monthly Passengers in Flights")
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=45)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = sns.load_dataset('iris')

# Violin plot for sepal width by species
sns.violinplot(x='species', y='sepal_width', data=df, palette='muted')

plt.title("ðŸŽ» Sepal Width by Species (Violin Plot)")
plt.xlabel("Species")
plt.ylabel("Sepal Width")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = sns.load_dataset('iris')

# Strip plot for petal length by species
sns.stripplot(x='species', y='petal_length', data=df, jitter=True, color='blue', size=7)

plt.title("ðŸ”˜ Petal Length by Species (Strip Plot)")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = sns.load_dataset('iris')

# Swarm plot for petal width by species
sns.swarmplot(x='species', y='petal_width', data=df, palette='Set2', size=8)

plt.title("ðŸ Petal Width by Species (Swarm Plot)")
plt.xlabel("Species")
plt.ylabel("Petal Width")
plt.show()
```
