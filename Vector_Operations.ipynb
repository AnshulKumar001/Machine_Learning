```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set()   # yahi magic line hai
```

**1. Plotting a vector**

```python
plt.quiver(0,0,4,5,angles='xy',scale_units='xy',scale=1,color='r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,-3,-8,angles='xy',scale_units='xy',scale=1,color='b')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

**2. Addition of two Vectors**

```python
vector1=np.asarray([0,0,2,3])
vector2=np.asarray([0,0,3,-2])
sum=vector1+vector2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**3. Multiplying a vector by a scaler**

```python
vector1=np.asarray([0,0,2,3])
vector2=2*vector1
print(vector2)
```

```python
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector1=np.asarray([0,0,2,3])
vector2=-0.5*vector1
print(vector2)
```

```python
plt.quiver(-0,-0,-1,-1.5, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**4. Dot Products**

```python
a=np.asarray([2,3])
b=np.asarray([4,4])
dot_product=np.dot(a,b)
print(dot_product)
```

**5. cross product**

```python
a=np.asarray([2,3])
b=np.asarray([4,4])
cross_product=np.cross(a,b)
print(cross_product)
```

**Projection**

```python
a=np.asarray([2,5])
v=np.asarray([8,-6])
projection=np.dot(a,v)/np.dot(v,v)*v
print(projection)
```
