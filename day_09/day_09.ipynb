# **Feature Scaling - Normalization**

---



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/wine_data.csv')
df
```

```python
df=pd.read_csv('/content/wine_data.csv',header=None,usecols=[0,1,2],
               names=['class label','Alcohol','Malic acid'])
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict={1:'red',2:'blue',3:'green'}
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['class label'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('class label',axis=1),
                                               df['class label'],
                                               test_size=0.3,
                                               random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))
ax1.scatter(X_train['Alcohol'], X_train['Malic acid'], c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic acid'], c=y_train)
ax2.set_title("After Scaling")
plt.show()

```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'], ax=ax1)
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))
# before scaling
ax1.set_title("Malic acid Distribution Before Scaling")
sns.kdeplot(X_train['Malic acid'], ax=ax1)
#after scaling
ax2.set_title("Malic acid Distribution After Scaling")
sns.kdeplot(X_train_scaled['Malic acid'],ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))
# before scaling
ax1.set_title("Alchohol Distribution Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax=ax1)
#after scaling
ax2.set_title("Alcohol Distribution After Scaling")
sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)
plt.show()
```
